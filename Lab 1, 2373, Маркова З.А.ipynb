```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('world_bank_dataset.csv')
df.head()
```

```python
def findOutliers(arr):
    Q1 = np.percentile(arr, 25, method='midpoint')
    Q3 = np.percentile(arr, 75, method='midpoint')
    IQR = Q3 - Q1
    
    
    bottom = Q1-1.5*IQR
    top = Q3+1.5*IQR
    print('Нижняя граница нормы: ',bottom,' Верхняя граница нормы: ',top)
    outliers = np.array((arr >= top) | (arr <= bottom))
    print("Количество выбросов:", outliers.sum())
```

```python
#Год
Year = np.array(df['Year'])
avgYear = Year.mean()
print('Среднее значение: ' + str(avgYear))

yearSKO = np.std(Year)
print('СКО: ' + str(yearSKO))

gap=df['Year'].isna().sum()
print('Количество пропусков: ' + str(gap))

plt.figure(figsize=(10, 5))
plt.hist(df['Year'], bins=30);
plt.grid(True)
plt.title('Year distribution')
plt.show()

findOutliers(Year)
```

```python
#ВВП
GDP = np.array(df['GDP (USD)'])
avgGDP = GDP.mean()
print('Среднее значение: ' + str(avgGDP))

gdpSKO = np.std(GDP)
print('СКО: ' + str(gdpSKO))

gap=df['GDP (USD)'].isna().sum()
print('Количество пропусков: ' + str(gap))

plt.figure(figsize=(10, 5))
plt.hist(df['GDP (USD)'], bins=30);
plt.grid(True)
plt.title('GDP distribution')
plt.show()

findOutliers(GDP)
```

```python
#Население
Pop = np.array(df['Population'])
avgPop = Pop.mean()
print('Среднее значение: ' + str(avgPop))

popSKO = np.std(Pop)
print('СКО: ' + str(popSKO))

gap=df['Population'].isna().sum()
print('Количество пропусков: ' + str(gap))

plt.figure(figsize=(10, 5))
plt.hist(df['Population'], bins=30);
plt.grid(True)
plt.title('Population distribution')
plt.show()

findOutliers(Pop)
```

```python
#Ожид. прод. жизни
Life = np.array(df['Life Expectancy'])
avgLife = Life.mean()
print('Среднее значение: ' + str(avgLife))

lifeSKO = np.std(Life)
print('СКО: ' + str(lifeSKO))

gap=df['Life Expectancy'].isna().sum()
print('Количество пропусков: ' + str(gap))

plt.figure(figsize=(10, 5))
plt.hist(df['Life Expectancy'], bins=30);
plt.grid(True)
plt.title('Life Expectancy distribution')
plt.show()

findOutliers(Life)
```

```python
#Безработица
Unemploy = np.array(df['Unemployment Rate (%)'])
avgUnemploy = Unemploy.mean()
print('Среднее значение: ' + str(avgUnemploy))

unemploySKO = np.std(Unemploy)
print('СКО: ' + str(unemploySKO))


gap=df['Unemployment Rate (%)'].isna().sum()
print('Количество пропусков: ' + str(gap))

plt.figure(figsize=(10, 5))
plt.hist(df['Unemployment Rate (%)'], bins=30);
plt.grid(True)
plt.title('Unemployment Rate (%) distribution')
plt.show()

findOutliers(Unemploy)
```

```python
#Углекислый газ
CO2 = np.array(df['CO2 Emissions (metric tons per capita)'])
avgCO2 = CO2.mean()
print('Среднее значение: ' + str(avgCO2))

co2SKO = np.std(CO2)
print('СКО: ' + str(co2SKO))

gap=df['CO2 Emissions (metric tons per capita)'].isna().sum()
print('Количество пропусков: ' + str(gap))

plt.figure(figsize=(10, 5))
plt.hist(df['CO2 Emissions (metric tons per capita)'], bins=30);
plt.grid(True)
plt.title('CO2 Emissions distribution')
plt.show()

findOutliers(CO2)
```

```python
#Электричество
Electr = np.array(df['Access to Electricity (%)'])
avgElectr = Electr.mean()
print('Среднее значение: ' + str(avgElectr))

electrSKO = np.std(Electr)
print('СКО: ' + str(electrSKO))

gap=df['Access to Electricity (%)'].isna().sum()
print('Количество пропусков: ' + str(gap))

plt.figure(figsize=(10, 5))
plt.hist(df['Access to Electricity (%)'], bins=30);
plt.grid(True)
plt.title('Access to Electricity (%) distribution')
plt.show()

findOutliers(Electr)
```
